```python
# tokenization
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
vocab_size = 10000
tokenizer = Tokenizer(num_words= vocab_size)
tokenizer.fit_on_texts(quotes)
```

```python
word_index = tokenizer.word_index
print(len(word_index))
list(word_index.items())[:10]
```

```python
sequence = tokenizer.texts_to_sequences(quotes)
```

```python
X = []
y = []
for seq in sequence:
    for i in range(1, len(seq)):
        input_seq = seq[:i]
        output_seq = seq[i]
        X.append(input_seq)
        y.append(output_seq)
```

```python
max_len = max(len(x) for x in X)
print(max_len)
```

```python
# padding
from tensorflow.keras.preprocessing.sequence import pad_sequences
X_padded = pad_sequences(X, maxlen = max_len, padding = "pre")
```

```python
# one -hot encoding on y
from tensorflow.keras.utils import to_categorical
y_one_code = to_categorical(y, num_classes= vocab_size)
```

```python
# embedding/ model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
embedding_dim = 50
rnn_units = 128
```

```python
rnn_model = Sequential()

rnn_model.add(
    Embedding(input_dim = vocab_size, output_dim= embedding_dim, input_length = max_len)
)
rnn_model.add(
    SimpleRNN(units= rnn_units,activation= "tanh")
)
rnn_model.add(Dense(units = vocab_size, activation= "softmax"))
```

```python
rnn_model.compile(
    optimizer= "adam",
    loss = "categorical_crossentropy",
    metrics= ["accuracy"]
)
```

```python
rnn_model.summary()
```

```python
# LSTM Model
from tensorflow.keras.layers import LSTM
lstm_model = Sequential()
lstm_model.add(
    Embedding(input_dim = vocab_size, output_dim = embedding_dim, input_length = max_len)
)
lstm_model.add(LSTM(units = rnn_units))
lstm_model.add(Dense(units= vocab_size, activation= "softmax"))
```

```python
lstm_model.compile(
    optimizer= "adam",
    loss= "categorical_crossentropy",
    metrics= ["accuracy"]
)
```

```python
lstm_model.summary()
```

```python
# GSU Model
from tensorflow.keras.layers import GRU
gru_model = Sequential()
gru_model.add(
    Embedding(input_dim = vocab_size, output_dim = embedding_dim, input_length = max_len)
)
gru_model.add(GRU(units = rnn_units))
gru_model.add(Dense(units= vocab_size, activation= "softmax"))
```

```python
gru_model.compile(
    optimizer= "adam",
    loss= "categorical_crossentropy",
    metrics= ["accuracy"]
)
```

```python
gru_model.summary()
```

```python
index_to_word = {}
for word, index in word_index.items():
    index_to_word[index] = word
```

```python
index_to_word
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
def predictor(model, tokenizer, text, max_len):
    text = text.lower()

    seq = tokenizer.texts_to_sequences([text])[0]
    seq = pad_sequences([seq], maxlen= max_len, padding= "pre")

    pred = model.predict(seq, verbose = 0)
    pred_index = np.argmax(pred)
    return index_to_word[pred_index]
```

```python
seed_text = "what are you "
next_word = predictor(lstm_model, tokenizer, seed_text, max_len)
print(next_word)
```

```python
def generate_text(model, tokenizer, seed_text, max_len, n_words):
    for _ in range(n_words):
        next_word = predictor(model, tokenizer, seed_text, max_len)
        if next_word == "" :
            break
        seed_text += " " + next_word
    return
```

```python
seed = " the meaning of life"
generate_text = generate_text(lstm_model, tokenizer, seed, max_len, 10)
print(generate_text)
```
